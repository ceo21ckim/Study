```python
# 1
import random

def stock_function_1(stock_price, _time):
    stock_prices = []
    stock_prices.append(stock_price)
    
    for _ in range(1, _time):
        random.seed(1234)
        v = random.uniform(0, 0.2)
        if stock_price*(1.2) >= stock_price * (1+v):
            if stock_price * v < 100 : pass
            else:
                stock_prices.append(stock_price * (1+v))
                stock_price = stock_prices[-1]

        else:
            stock_prices.append(stock_prices[-1])
            stock_price = stock_prices[-1]

    return stock_prices
```

```python
# 2
def stock_function_2(stock_price, _time):
    stock_prices = [stock_price]
    for _ in range(1, _time):
        random.seed(1234)
        v = random.uniform(100, stock_price*0.2)
        stock_prices.append(stock_price  + v) ; stock_price = stock_prices[-1]
        
    return stock_prices
```

```python
_time = 15-9+1
stock_price = 10000


stock1= stock_function_1(stock_price, _time)

stock2 = stock_function_2(stock_price, _time)
```

```python
import matplotlib.pyplot as plt
import numpy as np 

# plt.rcParams['font.family'] = 'Nanum Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
fig, ax = plt.subplots(figsize = (8, 5))
x_aes = [str(x+9)+':00' for x in range(_time)]
ax.plot(stock1, label = 'stock1', marker = 'o', linestyle = '--')
ax.plot(stock2, label = 'stock2', marker = 'o', linestyle = '--')
ax.set_xticks(np.arange(_time))
ax.set_xticklabels(x_aes)
ax.legend(loc = 'lower center', fontsize = 15)
plt.show()
```
