```python
# 현재 directory를 확인
os.getcwd()

# dataset이 있는 directory에 접근
os.chdir(path)

# load file
file = r'starbucks_dataset.csv'
starbucks = pd.read_csv(file)
```

```python
starbucks.plot()
plt.show()
```

```python
seoul = starbucks[starbucks['시도'] == '서울시'].groupby(['시군구']).count()['매장명']
```

```python
plt.figure(figsize =(7, 7), dpi = 100)
plt.plot(seoul.sort_values(ascending=False))
plt.xticks(np.arange(len(seoul)), label = seoul.sort_values(ascending=False).index, rotation = 40)
plt.show()
```

```python
plt.figure(figsize = (7, 7))
seoul.plot(kind ='pie' , rotatelabels=42)
plt.xticks(rotation = 50)
# plt.savefig('pie chart.png')
plt.show()
```

```python
plt.pie(seoul, labels = seoul.index, rotatelabels=43)
# plt.xticks(seoul.index)
plt.show()
```

```python
plt.figure(figsize = (7, 7))
seoul.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize = (7, 7))
seoul.plot(kind='hist', bins = 50)
plt.show()
```

```python
plt.figure(figsize = (7, 7))
seoul.plot(kind='density')
plt.show()
```
