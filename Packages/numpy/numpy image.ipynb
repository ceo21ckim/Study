```python
x = np.arange(-1, 3*np.pi+2, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.hlines(0, -1, 3*np.pi+2)
plt.show()
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x) 

plt.plot(x, y_sin)
plt.plot(x, y_cos) 
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine','Cosine'])
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2, 2, 2)
plt.plot(x, y_cos)
plt.title('Cosine')

plt.show()
```

```python
import numpy as np 
from scipy.spatial.distance import pdist, squareform 
```

```python
x = np.array([
    [0, 1],
    [1, 0],
    [2, 0]
])

# 해당 값에 대해서 거리를 구하는데 squaredform 을 통해서 정사각행렬로 만들어주었다. 
d = squareform(pdist(x, 'euclidean')) # euclidean distance 
print(d)
```

```python
img = imread('lena_gray.gif')
img2 = img.copy()
row, col = img.shape
```

```python
# median을 쓰는 이유는 noise를 잡아주기 때문이다. 
for i in range(row):
    for j in range(1, col-1):
        img2[i,j] = statistics.median(img[i,j-1 : j+2])

plt.subplot(1, 2, 1)
plt.imshow(img, cmap = 'gray')

plt.subplot(1, 2, 2)
plt.imshow(img2, cmap = 'gray')

plt.show()
```

```python
import statistics
img = imread('lena_color.jpg')
img2 = img.copy()
row, col, channel = img2.shape
# median을 쓰는 이유는 noise를 잡아주기 때문이다. 
for k in range(channel):
    for i in range(row):
        for j in range(1, col-1):
            img2[i,j,k] = statistics.median(img[i,j-1 : j+2,k])

plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(img2)

plt.show()
```
